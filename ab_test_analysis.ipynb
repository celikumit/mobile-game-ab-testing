```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Using a stable, alternative public source for the Cookie Cats dataset
url = "https://raw.githubusercontent.com/yfsui/Mobile-Games-AB-Testing-Project/master/cookie_cats.csv"
df = pd.read_csv(url)

print("--- Dataset Summary ---")
print(df.head())
print("\n--- Group Assignments ---")
print(df['version'].value_counts())
```

```python
# Split groups
control_rounds = df[df['version'] == 'gate_30']['sum_gamerounds']
treatment_rounds = df[df['version'] == 'gate_40']['sum_gamerounds']

# Run the test
mw_stat, mw_p_val = stats.mannwhitneyu(control_rounds, treatment_rounds, alternative='two-sided')

print(f"Control (Gate 30) Median Rounds: {control_rounds.median()}")
print(f"Treatment (Gate 40) Median Rounds: {treatment_rounds.median()}")
print(f"Mann-Whitney U Test p-value: {mw_p_val}")
```

```python
# Create a contingency table for 7-day retention
retention_table = pd.crosstab(df['version'], df['retention_7'])
print("--- 7-Day Retention Counts ---")
print(retention_table)

# Calculate percentages for context
print("\n--- Retention Percentages ---")
print(df.groupby('version')['retention_7'].mean() * 100)

# Run Chi-Square Test
chi2, chi2_p_val, dof, expected = stats.chi2_contingency(retention_table)
print(f"\nChi-Square Test p-value: {chi2_p_val}")
```
